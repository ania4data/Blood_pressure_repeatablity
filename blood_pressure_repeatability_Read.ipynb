```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

X=[]
Y=[]
for line in open('bp_repeatability.dct'):
    
#    print(type(int(line.lstrip().split(' ')[0])))
    
    try:
        
        id,O_J_1,O_J_2,O_J_3,O_R_1,O_R_2,O_R_3,M_1,M_2,M_3=line.strip().split('  ')
        Y.append([float(O_J_1),float(O_J_2),float(O_J_3),float(O_R_1),float(O_R_2),float(O_R_3),float(M_1),float(M_2),float(M_3)])
        X.append([int(id),1])
#        print((int(line.strip().split('  ')[0])),(int(line.strip().split('  ')[1])),(int(line.strip().split('  ')[2])))        
    except:
        
        print ('Pass!')

X=np.array(X)  
Y=np.array(Y)        
```

```python
plt.plot(X[:,0],sorted(Y[:,0]),label='Observer J1')
plt.plot(X[:,0],sorted(Y[:,1]),label='Observer J2')
plt.plot(X[:,0],sorted(Y[:,2]),label='Observer J3')
plt.plot(X[:,0],sorted(Y[:,3]),label='Observer R1')
plt.plot(X[:,0],sorted(Y[:,4]),label='Observer R2')
plt.plot(X[:,0],sorted(Y[:,5]),label='Observer R3')
plt.plot(X[:,0],sorted(Y[:,6]),label='Machine1')
plt.plot(X[:,0],sorted(Y[:,7]),label='Machine2')
plt.plot(X[:,0],sorted(Y[:,8]),label='Machine3')
plt.legend()
plt.title('Offset between Machine and human Blood pressure measurment \n specially at mid range BP per patient')
```
